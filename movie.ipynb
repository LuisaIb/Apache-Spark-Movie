```python
joined = pd.merge(movie, genre, left_on="id", right_on="movie_id")
joined=joined.drop('movie_id', axis=1)
joined=joined.drop('title_y', axis=1)
joined=joined.drop('id_y', axis=1)
joined = joined.rename(columns={'title_x': 'title', 'id_x': 'id'})
joined.head(10)
```

```python
data = pd.merge(joined, recommendations, left_on="id", right_on="movie_id")
data = data.drop('movie_id', axis=1)
data.head(10)
```

```python
data['median_rating'] = data.groupby(['genre_name'])['vote'].agg('median')
data = data.assign(median_rating=3)
data
```
